```python
def countDecreasingRatings(ratings):
  periods = len(ratings)
  for index, rating in enumerate(ratings[:-1]):
    for sum in range(len(ratings[index:])-1):
      index += 1
      if rating-(sum+1) == ratings[index]:
        periods += 1
      else:
        break

  print(periods)
  
```

```python
ratings = [6, 5, 4, 2, 3, 1]#[4, 3, 5, 4, 3][2, 1, 3][4, 2, 3, 1]
countDecreasingRatings(ratings)
```

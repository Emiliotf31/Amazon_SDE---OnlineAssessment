# Question

Alexa is Amazon's virtual AI assistant. It makes it easy to set up your Alexa-enabled devices, listen to music, get weather updates, and much more. **The team is working on a new feature that evaluates the *aggregate temperature change* for a period based on the changes in temperature of previous and upcoming days.**

Taking the change in temperature data in *n* days, the ***aggregate temperature change*** evaluated on the i_th day is the maximum of the sum of the changes in temperature until the i_th day and the sum of the change in temperature in the next (*n-i*) days, with the i_th day temperature change included in both.

Given the temperature data of *n* days, find the maximum ***aggregate temperature change* evaluated among all the days.**

The *aggregate temperature change* on each day is evaluated as:

Captura.PNG

For day 1, there are no preceding days's information, but the day itself is included in the calcuation. Temperature changes = [6] for the *before* period.

For succeeding days, there are values [6,-2,5] and 6+(-2)+5=9. Again, the change fir the day 1is included. The maximum of 6 and 9.

For day 2, consider [6,2] -> 6+(-2)=4, and [-2,5] -> (-2)+5=3. The maximum of 3 and 4 is 4.

For day 3, consider [6,-2,5] -> 6+(-2)+5=9, and [5]. The maximum of 9 and 5 is 9.

The maximum *aggregate temperature change* is *max(9,4,9)=9*.

**Function description**

Complete the function *getMaxAggregateTemperatureChange* in the editor below.

*getMaxAggregateTemperatureChange* has the following parameters:

> *int tempChange[n]:* the temperature change data of *n* days


**Returns**
> *long*: the maximum *aggregate temperature change*

**Constrains**
- 1 <= *n* <= 10^5
- -10^9 <= *tempChange[i]* <= 10\^9 where, 1 <= *i* <= *n*

# Solution

````python
def getMaxAggregateTemperature(tempChange):
  max_tempChange = []
  for day in range(len(tempChange)):
    Last_tempChange = tempChange[:day+1]
    Next_tempChange = tempChange[day:]
    last_sum = 0
    next_sum = 0
    for last_temp in Last_tempChange:
      last_sum += last_temp
    for next_temp in Next_tempChange:
      next_sum += next_temp
    """print("Temperaturas anteirores: ", Last_tempChange)
    print("Temperaturas siguientes: ", Next_tempChange)
    print("Suma anterior: ",last_sum)
    print("Suma siguiente: ", next_sum)"""
    max_tempChange_day = max(last_sum,next_sum)
    """print("Temperatura Máxima: ", max_tempChange_day)
    print("---------------------")"""
    max_tempChange.append(max_tempChange_day)
  print(max_tempChange)
  MATC = max(max_tempChange)
  print("Maximum aggregate temperature change: ", MATC)
  print("On day: ", max_tempChange.index(MATC)+1)  
````

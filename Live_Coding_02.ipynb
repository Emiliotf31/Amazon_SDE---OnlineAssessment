```python
def findMaximumSustainableClusterSize(processingPower, bootingPower, powerMax):
  k_s = []
  groups = []
  groups_indices = []
  group_index = 0
  for process in reversed(range(len(processingPower))):
    for index in range(len(processingPower[:process+1])):

      max_bootingPower = max(bootingPower[index:process+1])
      sum_processingPower = sum(processingPower[index:process+1])
      k = len(processingPower[index:process+1])
      net_power_consumption = max_bootingPower+sum_processingPower*k
      if net_power_consumption <= powerMax:
        k_s.append(k)

  if len(k_s) > 0:   
    print("Maxumim processors: ", max(k_s))
  else:
    print("0")
      
```

```python
bootingPower = [3, 6, 1, 3, 4]
processingPower = [2, 1, 3, 4, 5]
powerMax = 25

bootingPower = [8, 8, 10, 9, 12]
processingPower = [4, 1, 4, 5, 3]
powerMax = 33

findMaximumSustainableClusterSize(processingPower, bootingPower, powerMax)
```

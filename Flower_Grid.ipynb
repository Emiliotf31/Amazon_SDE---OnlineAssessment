# **Flower Grid**
# Question

While strolling in a garden, you come across a pattern of flowers. It goes as folows:
- The shape of the pattern is a square.
- The pattern consists of vairety of flowers. Let's represent each type of flower by an alphabet.
- If a single flower type is used, then we use 'A' to represent the flower and the pattern is a 2*2 grid with all A's:

```
AA
AA
```

# Solution

````python
import math
import numpy as np
````

````python
def flower_types(n_flowers):
  ls_flowers = []
  for flower in range(n_flowers):
    flower_type = chr(flower+65)
    ls_flowers.append(flower_type)
  return ls_flowers

def matrix(n_flower):
  ls_flowers = flower_types(n_flower)

  size_m = (len(ls_flowers)*2)
  help_m = np.full((size_m,size_m), [' '],dtype=str)

  max = size_m
  skip = size_m-1
  min = 0

  for elements in reversed(range(len(ls_flowers))):
    help_m[min:max:skip,min:max] = ls_flowers[elements]
    help_m[min:max,min:max:skip] = ls_flowers[elements]
    max -= 1
    skip -= 2
    min += 1

  print("Here is the pattern: ")
  print(help_m)

def user_define():
  n_flower = input("How many types of flowers did you find? ")
  matrix(int(n_flower))
````
